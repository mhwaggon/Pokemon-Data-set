```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


%matplotlib inline
sns.set(color_codes=True)
```

```python
pokemon = pd.read_csv('pokemon.csv', index_col=0)
```

```python
pokemon.head()
```

```python
pokemon.dtypes
```

```python
pokemon['type2'].fillna(value='', inplace=True)
pokemon.head()
```

```python
poketype = []
list1 = pokemon['type1']
list2 = pokemon['type2']
poketype = [x +' ' + y for x, y in zip(list1, list2)]
poketype = [z.rstrip() for z in poketype]
```

```python
pokemon['poketype'] = poketype
pokemon.tail()
```

```python
g = sns.countplot(x='generation', hue = 'is_legendary', data=pokemon)
g.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15),  shadow=True, ncol=2, labels=['NON-LEGENDARY','LEGENDARY'])
```

```python
g = pokemon.groupby(['generation']).mean()[['sp_attack', 'sp_defense']]
g.plot.bar(stacked = True)
```

```python
g = pokemon.groupby(['is_legendary']).mean()[['sp_attack', 'sp_defense', 'attack', 'defense']]
g.plot.bar(stacked = True)
```

```python
h = pokemon['is_legendary']!=0
g = pokemon.groupby([h]).mean()[['base_total']]
g.plot.bar(stacked = True)
```

```python
h = pokemon['is_legendary']!=0
g = pokemon.groupby([h, 'generation']).mean()[['attack', 'defense', 'sp_attack', 'sp_defense', 'hp']]
g.plot.bar(stacked = True)
```

```python
sns.catplot(
    y='poketype',
    data=pokemon,
    kind='count',
    order=pokemon['poketype'].value_counts().index,
    height=50,
    aspect=.25,
)
```

```python
plt.subplots(figsize=(15, 15))
sns.heatmap(
    pokemon[pokemon['type2']!=''].groupby(['type1', 'type2']).size().unstack(),
    linewidths=1,
    annot=True,
    cmap="Blues"
)

plt.xticks(rotation=35)
plt.show()
```
